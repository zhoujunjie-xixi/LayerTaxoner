```python
import pandas as pd
import numpy as np
import wget
import os
```

```python
p_dirname = 'K:/dataset/ncbi/pathogens'
csvname = 'Bocaparvovirus'
path = os.path.join(p_dirname, csvname)
```

```python
def download(url, path):
    if not os.path.exists(path):
        os.mkdir(path)
    wget.download(url, path)
```

```python
def get_url(ftp):
    url = ftp + '/' + ftp.split('/')[-1] + '_genomic.fna.gz'
    url = url.replace('ftp://', 'https://')
    return url
```

```python
df = pd.read_csv(csvname + '.csv')
df
```

```python
df_ref = df.dropna(subset=['RefSeq FTP'])
df_ref
```

```python
ser_ftp = df_ref['RefSeq FTP']
```

```python
for i in ser_ftp.index:
    print(ser_ftp[i])
    url = get_url(ser_ftp[i])
    #download(url, path)
    print(url)
```
